```python
Q1=train['BILL_AMT'].quantile(0.25)
Q3=train['BILL_AMT'].quantile(0.75)
IQR=Q3-Q1
train['BILL_AMT'] = train[~((train['BILL_AMT']< (Q1 - 1.5*IQR))| (train['BILL_AMT'] >(Q3 +1.5*IQR)))]
```

```python
Q1=train['QTY'].quantile(0.25)
Q3=train['QTY'].quantile(0.75)
IQR=Q3-Q1
train['QTY'] = train[~((train['QTY']< (Q1 - 1.5*IQR))| (train['QTY'] >(Q3 +1.5*IQR)))]
```

```python
Q1=train['VALUE'].quantile(0.25)
Q3=train['VALUE'].quantile(0.75)
IQR=Q3-Q1
train['VALUE'] = train[~((train['VALUE']< (Q1 - 1.5*IQR))| (train['VALUE'] >(Q3 +1.5*IQR)))]
```

```python
Q1=train['PRICE'].quantile(0.25)
Q3=train['PRICE'].quantile(0.75)
IQR=Q3-Q1
train['PRICE'] = train[~((train['PRICE']< (Q1 - 1.5*IQR))| (train['PRICE'] >(Q3 +1.5*IQR)))]
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)#Initialy we chose cluster=5
kfit = km.fit(train_scaled)
```

```python
k_pred =  kfit.predict(train_scaled)
k_pred
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(train)
    wcss.append(km.inertia_)
```
